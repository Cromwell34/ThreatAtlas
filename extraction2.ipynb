```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import requests
import json
import plotly.graph_objs as go
from datetime import datetime, timedelta
import time

app = dash.Dash()

# Specify the date range for the data you want to download
start_date = (datetime.now() - timedelta(days=1)).strftime('%Y%m%d')
end_date = datetime.now().strftime('%Y%m%d')

# Specify the parameters for the API request
params = {
    "query": "YOUR_QUERY",  # replace with your query
    "format": "JSON", 
    "timespan": '1M'
}

# Make the initial API request and parse the JSON response
response = requests.get("https://api.gdeltproject.org/api/v2/doc/doc", params=params)
data = json.loads(response.text)

app.layout = html.Div([
    dcc.Graph(id='gdelt-graph'),
    dcc.Interval(
        id='interval-component',
        interval=900000, # 15 minutes in milliseconds
        n_intervals=0
    )
])

@app.callback(Output('gdelt-graph', 'figure'),
              [Input('interval-component', 'n_intervals')])
def update_graph(n):
    # Make the API request and parse the JSON response
    response = requests.get("https://api.gdeltproject.org/api/v2/doc/doc", params=params)
    data = json.loads(response.content)

    # Create a trace for the graph
    trace = go.Scatter(
        x=[d['date'] for d in data],
        y=[d['value'] for d in data],
        mode='lines+markers',
        name='GDELT data'
    )

    return {'data': [trace], 'layout': go.Layout(title='GDELT data')}

if __name__ == '__main__':
    app.run_server(debug=True)
```
